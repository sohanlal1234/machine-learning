```python
avg_rd_by_state = df.groupby("State")["R&D Spend"].mean()
```

```python
avg_rd_by_state["California"]
```

```python
def replace_rd(row):
    rd = row["R&D Spend"]
    state = row["State"]
    rd = avg_rd_by_state[state] if np.isnan(rd) else rd
    return rd

df["R&D Spend"] = df.apply(replace_rd, axis = 1)
```
