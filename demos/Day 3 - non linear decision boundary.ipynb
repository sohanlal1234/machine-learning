```python
import pandas as pd
from sklearn import *
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("/data/iris.csv")
```

```python
df.head()
```

```python
df.Species.value_counts()
```

```python
features = ["SepalLengthCm", "PetalLengthCm"]
```

```python
fig, ax = plt.subplots()
colors = ["red", "green", "blue"]
for i, v in enumerate(df.Species.unique()):
    df[df.Species == v][features].plot.scatter(0, 1, ax = ax, color = colors[i], label = v)
plt.legend()
```

```python
from mlxtend.plotting.decision_regions import plot_decision_regions
```

```python
y = np.where(df.Species == "Iris-setosa", 1, 0)
X = df[features].values

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y
                                        , test_size = 0.3, random_state = 1)

pipe = pipeline.Pipeline([
    #("poly", preprocessing.PolynomialFeatures(degree=2, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LogisticRegression(random_state = 241, solver = "liblinear"))
])

pipe.fit(X_train, y_train)

est = pipe.steps[-1][-1]

y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)

print("Confusion matrix:\n", metrics.confusion_matrix(y_test, y_test_pred))
print("Test accuracy:", metrics.accuracy_score(y_test, y_test_pred))
print("Training accuracy:", metrics.accuracy_score(y_train, y_train_pred))

plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, pipe, X_highlight = X_test)
```
